# Collaborative Filtering - Streaming Algorithm

```python
#randomly generate User Matrix, Item Matrix & Initialize Rating Matrix
def generateRandUVR():
   U  =  np.random.uniform(-1,1,size = (k,n))# 5x12
   V  =  np.random.uniform(-1,1,size = (k,m))# 5x60
   R =   np.zeros((12,60))
   return U,V,R
```

```python
#Global variable
k = 5
n = 12
m = 60
R = np.zeros((n,m))
d=3
U,V,R = generateRandUVR()
alpha=0.1
lamda=0.1

def getVFromMatrixStore(p):
 return V[:,int((p)*m/d):int((p+1)*m/d)]

def setVinMatrixStore(q,Vq):
  V[:,int((q)*m/d):int((q+1)*m/d)] = Vq

def getUFromMatrixStore(p):
 return U[:,int((p)*n/d):int((p+1)*n/d)]

def setUinMatrixStore(p,Up):
  U[:,int(p*n/d):int((p+1)*n/d)] = Up
```

```python

def worker(i,j,r,p,q):
    R[i][j] = r # Store User i Item j Rating in R 
    UP =  getUFromMatrixStore(p) # 5x4
    VQ =  getVFromMatrixStore(q) # 5x20
    ui = i%4 
    vj = j%20

  #DSGD
    UP[:,ui] = UP[:,ui] - 2 * alpha * ((np.dot(UP[:,ui].T,VQ[:,vj])-R[i][j])*VQ[:,vj]+(lamda/m)*UP[:,ui]) 
    VQ[:,vj] = VQ[:,vj] - 2 * alpha * ((np.dot(UP[:,ui].T,VQ[:,vj])-R[i][j])*UP[:,ui]+(lamda/n)*VQ[:,vj])
  
  #save into matrix store
    setUinMatrixStore(p,UP) #save user matrix to matrix store
    setVinMatrixStore(q,VQ) #save item matrix to matrix store
```

```python
#Master algorithm that assigns process to worker and blocks same row and same column.
def master():
  for i in range(12):
    for j in range(60):
      p = math.floor(i/4)
      q = math.floor(j/20)
      r =  np.random.randint(0,6)
      worker(i,j,r,p,q) #initiates worker for the process

#Call master
master()      

#calculate MSE
print('mean squared error:',metrics.mean_squared_error(y_true = R,y_pred = np.dot(U.T,V)))
print('mean absolute error', metrics.mean_absolute_error(y_true = R, y_pred = np.dot(U.T,V)))
```
